```python
import numpy as np
from scipy.stats import mode

def knn(X_train, y_train, X_test, k):
    if isinstance(X_train, pd.DataFrame):
        X_train = X_train.to_numpy()
    if isinstance(X_test, pd.DataFrame):
        X_test = X_test.to_numpy()

    distances = np.linalg.norm(X_test[:, np.newaxis] - X_train, axis=2)
    nearest_neighbors = np.argsort(distances, axis=1)[:, :k]
    nearest_neighbor_labels = y_train.values[nearest_neighbors]

    # Use np.unique to find the mode
    predicted_labels = np.array([np.unique(x, return_counts=True)[0][np.argmax(np.unique(x, return_counts=True)[1])] for x in nearest_neighbor_labels])

    return predicted_labels
```

```python
import pandas as pd

iris_df = pd.read_csv("/content/drive/MyDrive/dsml_lab_dataset/iris.csv")
```

```python
X_train = pd.DataFrame(iris_df[["sepal.length", "sepal.width", "petal.length", "petal.width"]]).iloc[4:10, :]
y_train = iris_df["variety"].iloc[4:10]
X_test = pd.DataFrame(iris_df[["sepal.length", "sepal.width", "petal.length", "petal.width"]]).iloc[10:14]
y_test = iris_df["variety"].iloc[10:14]
```

```python
predicted_labels = knn(X_train, y_train, X_test, k=3)
accuracy = np.sum(predicted_labels == y_test) / len(y_test)
print('Accuracy:', accuracy)
```
