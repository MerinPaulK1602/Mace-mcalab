```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
df["diagnosis"]=label_encoder.fit_transform(df['diagnosis'])
```

```python
df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
train,test=train_test_split(df, test_size=0.3)
```

```python
trainX=train[df.columns[2:-1]]
trainy=train[df.columns[1]]
```

```python
testX=test[df.columns[2:-1]]
testy=test[df.columns[1]]
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(trainX.values, trainy.values)
```

```python
y_pred = gnb.predict(testX.values)
```

```python
print("model accuracy:", metrics.accuracy_score(testy.values, y_pred)*100)
```
