```python
#import numpy package
import numpy as np
```

```python
#import iris dataset
from sklearn.datasets import load_iris
```

```python
#importing libraries and packages for
#spliiting of dataset into for testing and training
from sklearn.model_selection import train_test_split
#model building
from sklearn.neighbors import KNeighborsClassifier
#accuracy
from sklearn.metrics import accuracy_score
```

```python
#loading dataset
iris=load_iris()
```

```python
#splitting into x and y variables
x=iris.data
y=iris.target
```

```python
#test and train varibles initialized
#test size fixed as 30%
#random state fixed as 42
X_train, X_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#calculating k value
#applying k value into model
k=int(np.sqrt(len(x)))
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
#fitting model
knn.fit(x,y)
```

```python
#sample test data
s=[[5.9, 2, 4.9, 1.9]]
```

```python
#prediction using model
pred=knn.predict(s)
print("Predicted Class:", iris.target_names[pred[0]])
```
