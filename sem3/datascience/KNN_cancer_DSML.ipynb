```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

importing libraries

```python
df.isnull().sum()
```

null values present for unnamed

```python
#handling missing data
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

all null values are corrected

```python
df['diagnosis'].unique()
```

identifying unique diagnosis values

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df["diagnosis"] = label_encoder.fit_transform(df['diagnosis'])
# 0 - B, 1 - M
df['diagnosis']
```

```python
df.columns
```

```python
train, test = train_test_split(df, test_size = 0.3)

trainX = train[df.columns[2:-1]]
trainY=train[df.columns[1]]

# trainY=trainY.astype('int')
testX= test[df.columns[2:-1]]
testY =test[df.columns[1]]
```

splitting test and train data
test proportion 30%

```python
c_knn = KNeighborsClassifier(n_neighbors=3)
c_knn.fit(trainX.values,trainY.values)

y_pred = c_knn.predict(testX.values)
print("Accuracy : ",metrics.accuracy_score(testY.values,y_pred))
```

calculating accuracy of model
94.74%
