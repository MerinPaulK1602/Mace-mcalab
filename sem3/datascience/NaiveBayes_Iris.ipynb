```python
from sklearn import preprocessing
```

```python
#data transformation
label_encoder = preprocessing.LabelEncoder()
df["variety"]=label_encoder.fit_transform(df['variety'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
train, test = train_test_split(df, test_size=0.3)
```

```python
trainX = train[df.columns[0:4]]
```

```python
trainX.shape
```

```python
print(trainX)
```

```python
trainy = train[df.columns[-1]]
```

```python
trainy.shape
```

```python
print(trainy)
```

```python
testX = test[df.columns[0:4]]
```

```python
testy = test[df.columns[-1]]
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(trainX.values, trainy.values)
```

```python
y_pred = gnb.predict(testX.values)
```

```python
print("model accuracy:", metrics.accuracy_score(testy.values, y_pred)*100)
```

```python
s = [[3.2,4.1,5.1,0.4]]
pred = gnb.predict(s)
res = pred[0]
if res == 0:
  print('Setosa')
if res == 1:
  print('Versicolor')
if res == 2:
  print('Virginica')
```

```python
s1 = [[5.9, 2, 4.9, 1.9]]
pred1 = gnb.predict(s1)
res1 = pred1[0]
if res1 == 0:
  print('Setosa')
if res1 == 1:
  print('Versicolor')
if res1 == 2:
  print('Virginica')
```
